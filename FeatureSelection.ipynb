<h1><center>Feature Selection</center></h1>
In this Notebook,  I will use [Boruta method](https://mbq.github.io/Boruta/) to select the main features.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.utils import shuffle

from boruta import boruta_py

import pandas as pd
import matplotlib.pyplot as plt
import copy

from caimcaim import CAIMD

from utility import random_forest_classifer

```

```python
data_train = pd.read_csv('data/train2.csv',index_col=0)
```

```python
data_train_happy=data_train.loc[data_train['TARGET']==0]
data_train_unhappy=data_train.loc[data_train['TARGET']==1]
s_perc = 0.5
data_train_happy_s = data_train_happy.sample(int(s_perc*data_train_unhappy.shape[0]),random_state=21)
data_train_unhappy_s= data_train_unhappy.sample(data_train_happy_s.shape[0],random_state=32)
data_train_homo = pd.concat([data_train_happy_s,data_train_unhappy_s])
```

```python
# define Boruta feature selection method
forest = RandomForestClassifier(n_estimators=100,max_features="sqrt",class_weight='balanced')

matrix_features = data_train_homo.as_matrix()[:,:-1]
labels = data_train_homo['TARGET'].as_matrix()


feat_selector = boruta_py.BorutaPy(forest, n_estimators='auto', verbose=2)
 
# find all relevant features
feat_selector.fit(matrix_features, labels)
 
# check selected features
features_s = feat_selector.support_
# check ranking of features
features_r = feat_selector.ranking_
```
